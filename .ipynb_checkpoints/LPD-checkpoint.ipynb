## 1 . train 이미지 데이터 h5 파일에 넣기

```python
base_path = './LPD_competition/'  # dataset path
save_path = './new-22.hdf5'  # path to save the hdf5 file

hf = h5py.File(save_path, 'a')  # open the file in append mode
train_label = []

for i in os.listdir(base_path):   # train
    vid_name = os.path.join(base_path, i)
    grp = hf.create_group(vid_name)  # create a hdf5 group.  each group is one 'A'

    for j in os.listdir(vid_name):  # 797
        track = os.path.join(vid_name, j)
        
        for h in range(48):
            train_label.append(j)
        subgrp = grp.create_group(j)  # create a subgroup for the above created group. each small
                                      # a is one subgroup

        for k in os.listdir(track):   # .jpg
            img_path = os.path.join(track, k)

            with open(img_path, 'rb') as img_f:  # open images as python binary
                binary_data = img_f.read()

            binary_data_np = np.asarray(binary_data)

            dset = subgrp.create_dataset(k, data=binary_data_np) # save it in the subgroup. each a-subgroup contains all the images.

hf.close()
```

## 3. train data 불러오기

```python
save_path = './new-19.h5'

data = []  # list all images files full path 'group/subgroup/b.png' for e.g. ./A/a/b.png. These are basically keys to access our image data.
group = [] # list all groups and subgroups in hdf5 file

def func(name, obj):     # function to recursively store all the keys
    
    if isinstance(obj, h5py.Dataset):
        data.append(name)
        
    elif isinstance(obj, h5py.Group):
        group.append(name)

hf = h5py.File(save_path, 'r')
hf.visititems(func)  # this is the operation we are talking about.


train_images_all = []

print(np.array(hf['0']))

# Now lets read the image files in their proper format to use it for our training.
cnt = 0
for j in data:
    kk = np.array(hf[j])
    img = np.array(PIL.Image.open(io.BytesIO(kk))) # our image file
    train_images_all.append(img)
    cnt += 1
    print('image size:', np.shape(img))

print(cnt)
```

## 4. train_valid_split

```python
train_images_all = np.array(train_images_all, dtype=int)
names_train = pd.Series(names_train)
```

```python
# Train : 60% Valid: 40% 나누기
train_images, valid_images, train_labels, valid_labels = train_test_split(train_images_all, names_train, test_size=0.4)
```

## 5. data generator

```python
datagen_kwargs = dict(rescale=1./255)
dataflow_kwargs = dict()

# Train
train_datagen = ImageDataGenerator(
        zoom_range=0.2,
        rotation_range = 20,
        brightness_range=[0.6,1.0],
    **datagen_kwargs
)

train_generator = train_datagen.flow(
    train_images, 
    y=train_labels,
    **dataflow_kwargs
)

# Validation
valid_datagen = ImageDataGenerator(**datagen_kwargs)
valid_generator = valid_datagen.flow(
    valid_images,
    y=valid_labels,
    shuffle=False, 
    **dataflow_kwargs
)

# # Test
# test_datagen = ImageDataGenerator(**datagen_kwargs)
# test_generator = test_datagen.flow(
#     test_images, 
#     y=names_test, 
#     shuffle=False, 
#     **dataflow_kwargs
# )
```

## 6. MLP 모델

```python
image_height = 256
image_width = 256
image_channel = 3
```

```python
model = Sequential([
    Input((image_height, image_width, image_channel)),
    Conv2D(filters=256, kernel_size=3, padding='same', activation='relu'),
    MaxPool2D(pool_size=(2, 2)),
    
    Conv2D(filters=512, kernel_size=3, padding='same', activation='relu'),
    MaxPool2D(pool_size=(2, 2)),
    
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(rate=0.25),
    Dense(43, activation='softmax')
])

model.summary()
```

```python
model.compile(
    loss=tf.keras.losses.sparse_categorical_crossentropy,
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9),
    metrics=['accuracy']
)
```

```python
# 처음 만든 모델이라면 EPOCHS를 1~5개로 하여 잘 돌아가는지 
# 성능을 확인해보고 값을 증가 시켜 봅시다. 
EPOCHS = 100

# EPOCHS에 따른 성능을 보기 위하여 history 사용
history = model.fit(
    train_images,
    train_labels, 
    validation_data = (valid_images, valid_labels),
    epochs=EPOCHS
)
```
